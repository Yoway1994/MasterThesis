```python
pump = [25, 60, 120, 204, 305, 430, 560, 710, 865, 1010, 1160, 1310, 1430, 1550, 1640, 1725, 1780]
RsP = [0.37, 2.1, 9.8, 31.6, 69.4, 125.5, 197.5, 274.6, 358.8, 442, 523.8, 601, 678.1, 740, 794, 835, 862]
Abs = [0.9852, 0.965, 0.918333333, 0.845098039, 0.772459016, 0.708139535, 0.647321429, 0.613239437, 0.585202312, 0.562376238, 0.548448276, 0.541221374, 0.525804196, 0.522580645, 0.515853659, 0.515942029, 0.515730337]

import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0, 1800)
y = 0.55*x - 120

sp = ['right', 'left', 'top', 'bottom']

fig1 = plt.figure('RsP')
ax1 = fig1.add_subplot(111)
ax1.tick_params(axis = 'y', direction = 'in')
ax1.tick_params(axis = 'x', direction = 'in')
ax1.set_ylabel('Residual pump power (mW)', fontsize = 14)
ax1.set_xlabel('Input pump power (mW)', fontsize = 14)
ax1.plot(pump, RsP, 'ks')
ax1.plot(x, y, 'b--')
ax1.set_ylim(-15)

ax2 = plt.twinx()
for i in sp:
    ax2.spines[i].set_linewidth(3)
ax2.spines['right'].set_color('red')
ax2.tick_params(axis = 'y', colors = 'red', direction = 'in')
ax2.set_ylabel('Absorbed pump ratio (a.u.)', fontsize = 14)
ax2.plot(pump, Abs, linestyle = '--', color = 'red', marker = 'o')
fig1.legend(loc = 1, labels = ['measurement', 'fitting line', 'absorbed ratio'], bbox_to_anchor=(0.6,1), bbox_transform = ax1.transAxes)

plt.show()
```
