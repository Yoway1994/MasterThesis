```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2* np.pi, 1001)
y = np.sin(x)
a = np.linspace(0, 2* np.pi, 20001)
l1 = np.linspace(2.2, 4.3, 18)
l2 = np.linspace(0, 1, 12)
l3 = np.linspace(5.5, 6.2, 8)
l = np.hstack((l1, l2, l3))
for i in l:
    K_e = 0.01
    b = np.exp(-((a-i)/(K_e*a))**2)
    plt.plot(a, b)

plt.plot(x, y, 'k')
#plt.show()

plt.savefig('D:\Desktop\KTN.png', format='png', bbox_inches='tight', transparent=True, dpi=600)
```
