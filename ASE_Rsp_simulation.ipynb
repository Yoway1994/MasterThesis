```python
def setting():
    ax = plt.gca()
    sp = ['right', 'left', 'top', 'bottom']
    for i in sp:
        ax.spines[i].set_color('black')
        ax.spines[i].set_linewidth(2)
    ax.tick_params(axis = "y", direction = "in")
    ax.tick_params(axis = "x", direction = "in")
```

```python
plt.figure('ASE')
#plt.plot(h['pump'], h['ASE'], 'r--', label = 'simulation')
#plt.plot(n['pump'], n['ASE'], 'b--', label = 'simulation')
plt.plot(n_pump_exp, n_ASE_exp, 'bo--', label = 'normal')
plt.plot(h_pump_exp, h_ASE_exp, 'rs--', label = 'high concentration')
setting()
plt.xlabel('Pump power (W)', fontsize = 14)
plt.ylabel('ASE power (μW)', fontsize = 14)
plt.xlim(0, 1)
plt.legend()

plt.figure('RsP')
#plt.plot(h['pump'], h['Rsp'], 'r--')
#plt.plot(n['pump'], n['Rsp'], 'b--')
plt.plot(n_pump_exp, n_RsP_exp, 'bo--', label = 'normal')
plt.plot(h_pump_exp, h_RsP_exp, 'rs--', label = 'high concentration')
setting()
plt.xlabel('Pump power (W)', fontsize = 14)
plt.ylabel('Residual pump power (W)', fontsize = 14)
plt.ylim(-0.02, 0.4)
plt.xlim(0, 1)
plt.legend()

plt.figure('Gain')
plt.plot(h['pump'], h['Gain'], 'r:', label = u'6.8x10^17 /cm\u00b2')
plt.plot(m['pump'], m['Gain'], 'b--', label = u'5.1x10^17 /cm\u00b2')
plt.plot(n['pump'], n['Gain'], 'k-', label = u'3.4x10^17 /cm\u00b2')

setting()
plt.xlabel('Pump power (W)', fontsize = 14)
plt.ylabel('Gain (dB)', fontsize = 14)
plt.ylim(-0.02, 2.5)
plt.xlim(0, 1.4)
plt.legend()
plt.show()
```
