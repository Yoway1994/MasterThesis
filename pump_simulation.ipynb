嘗試模擬出pump隨著HWP選轉的變化情形

```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
from numpy import pi
```

```python
path = r'D:\desktop\樣本資料\pump'
os.chdir(path)
file = os.listdir()[-1]
PER = pd.read_excel(file, dtype = np.float64)
```

```python
angles = np.linspace(0, 2*np.pi, 36, endpoint=False)
angles = np.concatenate((angles,[angles[0]]))
PER_data = np.concatenate((PER['power'], [PER['power'][0]]))
fig = plt.figure()
ax = fig.add_subplot(111, polar = True)
ax.plot(angles, PER_data, color = 'black')
ax.set_xticks(np.linspace(0,2*np.pi, 4, endpoint = False))
ax.set_rticks(np.linspace(0,600,7))
ax.set_rlabel_position(55)
ax.set_ylim(0, 550)
ax.legend(loc = 1, labels = ['pump power'],
         bbox_to_anchor=(1.2,1.1), bbox_transform = ax.transAxes)
plt.show()
```

```python
pump_exp = [111, 127, 167, 244, 357, 490, 650, 841, 1041, 1237, 1454, 1644, 1837, 
            2016, 2175, 2311, 2400, 2468]
pump_exp = (np.array(pump_exp) - 111)*1.2 + 111

pump_angle = np.linspace(0, 42.5, 18)
pump_a = np.linspace(0, 45, 19)
pump_angle = pump_angle - 0.8

angle = np.linspace(pi/2, 0, 19)
HWP = np.cos(angle) ** 2
max_p = 2849
pump_fit = max_p * HWP + 111
```

```python
plt.plot(pump_angle, pump_exp, 'ks')
plt.plot(pump_a, pump_fit, color = 'blue', linestyle = '--')

ax = plt.gca()
sp = ['right', 'left', 'top', 'bottom']
for i in sp:
    ax.spines[i].set_color('black')
    ax.spines[i].set_linewidth(2)
ax.tick_params(axis = "y", direction = "in")
ax.tick_params(axis = "x", direction = "in")
plt.xlabel('HWP angle (degree)')
plt.ylabel('transmitted pump power (mW)')
plt.legend(loc = 'best', labels = ['measurement', 'simulation'])

plt.show()
```
