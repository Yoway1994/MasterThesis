```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 建構一個二維資料
pump_f = 1
Pump_f = np.linspace(0, pump_f, 21)[np.newaxis, :]
z = np.linspace(0,10,11)[:, np.newaxis]
Ppf = Pump_f* np.exp(-z)
```

```python
print(Ppf[0,:])
print(Ppf[:, 0])
```

```python
print(np.shape(Ppf))
print(np.exp(-z))
plt.plot(z, Ppf)
plt.show()
```
