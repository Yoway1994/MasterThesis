```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
z = np.linspace(1, 1000) # mm, 透鏡和光纖出光處的距離 
f =  300 # mm, 透鏡焦距
Z_R = 75.1* 10**3 # mm, Rayleign range
M = abs(f/(z-f))* (1 + (Z_R/(z-f))**2)**(-1/2) # 光點放大倍率

w_0 = 10 # micron, 原始雷射光點大小 
w_spot = w_0* M # 通過透鏡後的光點大小

plt.plot(z, w_spot)
print(w_spot)
```
