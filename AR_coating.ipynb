```python
import os
path = r'D:\Desktop/NTU\碩士論文\coating data'
os.chdir(path)
file = os.listdir()[3]
print(os.listdir())
print(file)
```

```python
import pandas as pd
import numpy as np
columns = ['W', 'R', 'T', 'RP', 'TP']
D_AR = pd.read_excel(file, 'Sheet3')[4:]
D_AR = pd.DataFrame(D_AR, dtype = np.float64)
D_AR.columns = columns
M_AR = pd.read_excel(file, 'Sheet4')[4:]
M_AR = pd.DataFrame(M_AR, dtype = np.float64)
M_AR.columns = columns
```

```python
import matplotlib.pyplot as plt
ax = plt.gca()
sp = ['right', 'left', 'top', 'bottom']
for i in sp:
    ax.spines[i].set_color('black')
    ax.spines[i].set_linewidth(2)
ax.tick_params(axis = "y", direction = "in")
ax.tick_params(axis = "x", direction = "in")
plt.xlabel('Wavelength(nm)')
plt.ylabel('Reflectance(%)')
plt.plot(D_AR['W'], D_AR['R'], color = 'black')
plt.plot(M_AR['W'], M_AR['R'], color = 'blue', linestyle = '--')
plt.legend(loc = 'best', labels = ['2 layer AR', '1 layer AR'])
plt.show()
```
