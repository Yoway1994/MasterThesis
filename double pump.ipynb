```python
import matplotlib.pyplot as plt
import pandas as pd
import os

path = r'D:\desktop\碩士論文\double pump'
os.chdir(path)
file = os.listdir()
DP = pd.read_csv(file[0])
DPn2 = pd.read_csv(file[1])
SP = pd.read_csv(file[2])
SPn2 = pd.read_csv(file[-1])

plt.plot(DP['pump']*2, DP['Gain'], 'b--', label = 'DP')
plt.plot(SP['pump'], SP['Gain'], 'k-', label = 'SP')
ax = plt.gca()
sp = ['right', 'left', 'top', 'bottom']
for i in sp:
    ax.spines[i].set_color('black')
    ax.spines[i].set_linewidth(2)
ax.tick_params(axis = "y", direction = "in")
ax.tick_params(axis = "x", direction = "in")
plt.xlabel('Pump power (W)', fontsize = 14)
plt.ylabel('Gain(dB)', fontsize = 14)
plt.legend(loc = 'best')
plt.xlim(0,2)
plt.show()

plt.plot(DPn2['z']*10**3, DPn2['N2'], 'b--', label = 'DP')
plt.plot(SPn2['z']*10**3, SPn2['N2'], 'k-',label = 'SP')
ax = plt.gca()
sp = ['right', 'left', 'top', 'bottom']
for i in sp:
    ax.spines[i].set_color('black')
    ax.spines[i].set_linewidth(2)
ax.tick_params(axis = "y", direction = "in")
ax.tick_params(axis = "x", direction = "in")
plt.xlabel('Fiber position (mm)', fontsize = 14)
plt.ylabel('N\u2082/N\u209C (a.u.)', fontsize = 14)
plt.legend(loc = 'best')

plt.ylim(0.975, 0.991)
plt.show()
```
