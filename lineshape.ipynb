```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import exp, sqrt, pi
from sklearn.preprocessing import normalize
```

```python
x = np.linspace(1100, 1700, 601)
lambda_c = 1431
K_e = 0.134

sigma_e = (x**2)*exp(-((x-lambda_c)/(K_e*x))**2)
intensity = (x**(-3))*exp(-((x-lambda_c)/(K_e*x))**2)

plt.plot(x, sigma_e/np.max(sigma_e), 'red')
plt.plot(x, intensity/np.max(intensity), 'blue')
plt.plot(x, exp(-((x-lambda_c)/(K_e*x))**2), 'green')
plt.show()
```
