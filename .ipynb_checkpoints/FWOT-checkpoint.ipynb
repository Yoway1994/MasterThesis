```python
import os
path = r'D:\desktop\碩士論文\coating data'
os.chdir(path)
file = os.listdir()[-2]
print(os.listdir())
```

```python
import pandas as pd
import numpy as np
H = pd.read_excel(file, 'Sheet3')
L = pd.read_excel(file, 'Sheet4')
x = np.linspace(0, 2, 101)
y = 4.099526* np.ones(101)
```

```python
import matplotlib.pyplot as plt
plt.plot(H['Layer Thickness (FWOT)'], H['Reflectance (%)'], color = 'black', linestyle = ':')
plt.plot(L['Layer Thickness (FWOT)'], L['Reflectance (%)'], color = 'blue')
plt.plot(x, y, color = 'red', linestyle = '--')
ax = plt.gca()
sp = ['right', 'left', 'top', 'bottom']
for i in sp:
    ax.spines[i].set_color('black')
    ax.spines[i].set_linewidth(2)
ax.tick_params(axis = "y", direction = "in")
ax.tick_params(axis = "x", direction = "in")
plt.xlabel('Layer Thickness (FWOT)')
plt.ylabel('Reflectance (%)')
plt.legend(loc = 'best', labels = ['n = 2.1', 'n = 1.4', 'glass'])
plt.show()
```
