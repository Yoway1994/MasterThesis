```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2* np.pi, 1001)
y = np.sin(x)
a = np.linspace(0, 2* np.pi, 2001)
lambda_c = np.linspace(0.5, 2.5, 18)
for i in lambda_c:
    K_e = 0.01
    b = np.exp(-((a-i)/(K_e*a))**2)
    plt.plot(a, b)
lambda_l = np.linspace(3.8, 5.5, 16)
for i in lambda_l:
    K_e = 0.01
    g = np.exp(-((a-i)/(K_e*a))**2)
    plt.plot(a, g)
plt.plot(x, y, 'k')
#plt.show()

plt.savefig('D:\Desktop\KTN.png', format='png', bbox_inches='tight', transparent=True, dpi=600)
```
