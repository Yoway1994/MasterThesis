```python
b_w = [1409, 1420, 1422, 1442, 1446, 1466, 1475]
b_p = [10, 60, 120, 310, 300, 170, 50]
f_w = [1409, 1422, 1438, 1443, 1447, 1462.6, 1467.7]
f_p = [8.4, 15, 20, 31, 23.7, 17, 13.7]
```

```python
import numpy as np
lambda_c = 1447
x = np.linspace(1400, 1500, 101)
K_e = 0.016
g = np.exp(-((x-lambda_c)/(K_e*x))**2)
```

```python
import matplotlib.pyplot as plt
plt.plot(b_w, b_p/np.max(b_p), 'o')
plt.plot(f_w, f_p/np.max(f_p), 'o')
plt.plot(x, g/np.max(g))

plt.show()
```
