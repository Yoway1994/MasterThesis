```python
import os
path = r'D:\desktop\碩士論文\laser data'
os.chdir(path)
file = os.listdir()[0]
```

```python
import pandas as pd
data = pd.read_excel(file)
```

```python
import numpy as np
import matplotlib.pyplot as plt
angles = np.linspace(0, 2*np.pi, 36, endpoint=False)
angles = np.concatenate((angles,[angles[0]]))
data_200mW = np.concatenate((data['200mW'],[data['200mW'][0]]))
data_400mW = np.concatenate((data['400mW'],[data['400mW'][0]]))
data_800mW = np.concatenate((data['800mW'],[data['800mW'][0]]))

fig = plt.figure()
ax = fig.add_subplot(111, polar = True)
ax.plot(angles, data_200mW, color = 'red')
ax.plot(angles, data_400mW, color = 'blue')
ax.plot(angles, data_800mW, color = 'black')
ax.set_xticks(np.linspace(0,2*np.pi, 4, endpoint = False))
ax.set_rticks(np.linspace(0,70,5))
#ax.set_rticks([])
ax.set_rlabel_position(55)
ax.set_ylim(0,70)
ax.legend(loc = 1, labels = ['150mW', '300mW','600mW'],
         bbox_to_anchor=(0.3,1.1), bbox_transform = ax.transAxes)
plt.show()
```
