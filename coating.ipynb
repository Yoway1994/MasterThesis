```python
import pandas as pd
file_name = '鍍膜設計.xlsx'
HR = pd.read_excel(file_name, 'Sheet1')[4:]
AR = pd.read_excel(file_name, 'Sheet2')[4:]
new_columns = ['w', 'R%', 'T%', 'RP', 'TP']
HR.columns = new_columns
AR.columns = new_columns
```

```python
import matplotlib.pyplot as plt
plt.plot(HR['w'], HR['T%'], color = 'red', linestyle = '--')
plt.plot(AR['w'], AR['T%'], color = 'black')

ax = plt.gca()
sp = ['right', 'left', 'top', 'bottom']
for i in sp:
    ax.spines[i].set_color('black')
    ax.spines[i].set_linewidth(2)
ax.tick_params(axis = "y", direction = "in")
ax.tick_params(axis = "x", direction = "in")
plt.xlabel('Wavelength(nm)')
plt.ylabel('Transmittance(%)')
plt.legend(loc = 'lower left', labels = ['Input', 'Output AR'])

plt.show()
```
