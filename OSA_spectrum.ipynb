```python
import csv
import numpy as np 
import os
import matplotlib.pyplot as plt

path = r'D:\DeskTop\樣本資料\140523-2\tuning\800mW'
os.chdir(path)

def pure_val(csv_reader):
    list = []
    for i, rows in enumerate(csv_reader):
        if i > 28:
            np.seterr(divide = 'ignore')
            v = 10 * np.log10(abs(float(rows[1])))
            list.append(v)
    return list

def pure_wav(csv_reader):
    list = []
    for i, rows in enumerate(csv_reader):
        if i > 28:
            list.append(round(float(rows[0]),1))
    return list

total_x = []
total_y = []

for file in os.listdir():
    file_path = os.path.join(path, file)
    with open(file_path, 'r') as csv_file:
        csv_reader = csv.reader(csv_file)
        total_x.append(pure_wav(csv_reader))


for file in os.listdir():
    file_path = os.path.join(path, file)
    with open(file_path, 'r') as csv_file:
        csv_reader = csv.reader(csv_file)
        total_y.append(pure_val(csv_reader))

# total_y_norm = []		
		
# for i in total_y:
	# norm = [float(e)/max(i) for e in i]
	# total_y_norm.append(norm)
	
# print(total_y_norm)

for i in range(len(total_y)):
	plt.plot(total_x[i], total_y[i], lw = 0.5)

plt.ylim((-55, -10))
plt.xlim((1370, 1490))
plt.xticks(np.arange(1370, 1490, 20))
plt.show()
```
