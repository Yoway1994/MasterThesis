# Euler method

基本上前進的步幅越小(dx越小)，出來的結果就會更精確

下圖以 y = exp(y), y' = y 為例

給定初始值 y0 = 1 並用Euler method去取得近似曲線

```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import exp
def dydz(y):
    return y

L = 10 # total length
step = 1000 # the step of Euler method
dx = L/step # number of steps

y = np.zeros(step) # y columns
y[0] = 1 # initial value

for i in range(1, step):
    y[i] = y[i-1] + dx*dydz(y[i-1])

# x-axis
x = np.linspace(1, step, step, dtype = int)
X = dx*x

# plot
plt.plot(X, y, 'o') # fitting
plt.plot(X, exp(X)) 
plt.show()
```
