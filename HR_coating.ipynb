```python
import os
path = r'D:\desktop\NTU/碩士論文'
os.chdir(path)
file = os.listdir()[-1]
```

```python
import pandas as pd
HR = pd.read_excel(file)[4:]
HR.columns = ['W', 'R', 'T', 'RP', 'RT']
```

```python
import matplotlib.pyplot as plt
plt.figure('HR')
plt.plot(HR['W'], HR['R'], color = 'black')
ax = plt.gca()
sp = ['right', 'left', 'top', 'bottom']
for i in sp:
    ax.spines[i].set_color('black')
    ax.spines[i].set_linewidth(2)
ax.tick_params(axis = "y", direction = "in")
ax.tick_params(axis = "x", direction = "in")
plt.xlabel('Wavelength(nm)')
plt.ylabel('Reflectance(%)')
plt.legend(loc = 'best', labels = ['HR coating'])
plt.figure('HR_')
start = 470
end = 630
plt.plot(HR['W'][start:end], HR['R'][start:end], color = 'black')
ax = plt.gca()
sp = ['right', 'left', 'top', 'bottom']
for i in sp:
    ax.spines[i].set_color('black')
    ax.spines[i].set_linewidth(2)
ax.tick_params(axis = "y", direction = "in")
ax.tick_params(axis = "x", direction = "in")
plt.xlabel('Wavelength(nm)')
plt.ylabel('Reflectance(%)')
plt.ylim((97, 100))
plt.legend(loc = 'best', labels = ['HR coating'])
plt.show()
```
