```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

path = r'D:\desktop\NTU/碩士論文\concentration_sim'
os.chdir(path)
file = os.listdir()

p_20W = []
p_15W = []
p_10W = []
p_5W = []
for i in file:
    df = pd.read_csv(i)
    G = df['Gain']

    p_5W.append(G[1])
    p_10W.append(G[2])
    p_15W.append(G[3])
    p_20W.append(G[4])
x_c = np.linspace(1,13,13)*10**17
ax = plt.gca()
sp = ['right', 'left', 'top', 'bottom']
for i in sp:
    ax.spines[i].set_color('black')
    ax.spines[i].set_linewidth(2)
ax.tick_params(axis = "y", direction = "in")
ax.tick_params(axis = "x", direction = "in")
plt.plot(x_c, p_20W, 'k-', label = '10 W')
#plt.plot(x_c, p_15W, 'b:', label = '15W')
plt.plot(x_c, p_10W, 'r--', label = '5 W')
plt.plot(x_c, p_5W, 'b:', label = '2.5 W')

plt.xlabel('Cr\u2074\u207A ion concentration (cm\u207B\u00B3)', fontsize = 14)
plt.ylabel('Gain (dB)', fontsize = 14)
plt.legend(loc = 'best')
plt.show()
```
