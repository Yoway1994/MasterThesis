```python
import os
path = r'D:\desktop\碩士論文\coating data'
os.chdir(path)
file = os.listdir()[4]
```

```python
import pandas as pd
AR_exp = pd.read_excel(file, 'AR_exp')
AR_sim = pd.read_excel(file, 'AR_sim')
HR_exp = pd.read_excel(file, 'HR_exp')
HR_sim = pd.read_excel(file, 'HR_sim')
```

```python
import matplotlib.pyplot as plt
plt.figure('AR')
sp = ['right', 'left', 'top', 'bottom']
plt.plot(AR_exp['Wavelength (nm)'], AR_exp['Transmittance (%)'], color = 'black')
plt.plot(AR_sim['Wavelength (nm)'], AR_sim['Transmittance (%)'], color = 'blue', linestyle = '--')
ax = plt.gca()
for i in sp:
    ax.spines[i].set_color('black')
    ax.spines[i].set_linewidth(3)
ax.tick_params(axis = "y", direction = "in")
ax.tick_params(axis = "x", direction = "in")
plt.xlabel('Wavelength (nm)')
plt.ylabel('Transmittance (%)')
plt.legend(loc = 'best', labels = ['measurement', 'design'])

plt.figure('HR')
plt.plot(HR_exp['Wavelength (nm)'], HR_exp['Transmittance (%)'], color = 'black')
plt.plot(HR_sim['Wavelength (nm)'], HR_sim['Transmittance (%)'], color = 'blue', linestyle = '--')
ax = plt.gca()
for i in sp:
    ax.spines[i].set_color('black')
    ax.spines[i].set_linewidth(3)
ax.tick_params(axis = "y", direction = "in")
ax.tick_params(axis = "x", direction = "in")
plt.xlabel('Wavelength (nm)')
plt.ylabel('Transmittance (%)')
plt.legend(loc = 'lower left', labels = ['measurement', 'design'])

plt.show()
```
